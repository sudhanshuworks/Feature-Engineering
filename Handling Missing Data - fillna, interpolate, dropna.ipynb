```python
import pandas as pd
```

```python
df=pd.read_excel("weather_data.xlsx")
df
```

```python
df.info()
```

## fillna
**Fill all NaN with one specific value**

```python
new_df=df.fillna(0)  #in braket u can fill any value
new_df
```

#### **her ek column k liye different values **

```python
new_df=df.fillna({'temperature':1,'windspeed':3,'event':'no Event'})
new_df
```

```python
new_df['temperature'].dtype
```

### **Use method to determine how to fill na values**



```python
new_df=df.fillna(method='ffill')
new_df
```

```python
new_df = df.fillna(method="bfill")
new_df
```

### **Use of axis** column se next column me

```python
new_df=df.fillna(method='bfill',axis='columns')
new_df
```

### limit se next kitni row me same value paste hone h ye pata chalta h

```python
new_df=df.fillna(method='ffill',limit=1)
new_df
```

### dropna

```python
new_df=df.dropna()
new_df
```

```python
#jis column me sari value nan ho bus idex ho chor k wo row drop ho jati h
new_df=df.dropna(how='all')
new_df
```

```python
#thresh 1 ki wajah se wo row drop hogi jis row me 1 column me bhi value na ho
#thresh 2 krne per jis wo row drop hogi jisme 2 se kam column me value nai hogi
new_df=df.dropna(thresh=2)
new_df
```

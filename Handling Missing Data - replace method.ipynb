```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("temp_data.xlsx")
df
```

#### Replacing single value

```python
new_df=df.replace(-99999,value=0)
new_df
```

#### Replacing list with single value

```python
new_df=df.replace(to_replace=[-99999,0],value=22)
new_df
```

#### Replacing per column

```python
new_df = df.replace({
        'temperature': -99999,
        'windspeed': -99999,
        'event': '0'
    }, np.nan)
new_df
```

#### Replacing by using mapping

```python
new_df = df.replace({
        -99999: np.nan,
        'no event': 'Sunny',
    })
new_df
```

#### Regex

```python
#(magnitute/value and scale) so to remove scale or fix it use regex
# when windspeed is 6 mph, 7 mph etc. & temperature is 32 F, 28 F etc.
new_df = df.replace({'temperature': '[A-Za-z]', 'windspeed': '[a-z]'},'', regex=True) 
new_df
```
